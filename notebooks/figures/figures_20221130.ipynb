```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Translations
min_b = 0.01
max_b = 0.07
num_t = 200

beta = np.linspace(min_b, max_b, num_t)
alpha = 1 - beta
bar_alpha = np.cumprod(alpha)
```

```python
plt.plot(np.sqrt(bar_alpha), label=r'$\sqrt{\bar{\alpha}}$')
plt.plot(np.sqrt(1 - bar_alpha), label=r'$\sqrt{1 - \bar{\alpha}}$')
# plt.plot(np.sqrt(bar_alpha), label=r'$$\sqrt{\bar{\alpha}}$$')
plt.legend()
plt.title('Translation diffusion parameters')
plt.xlabel('t')
```

```python
# Rotations

# Rotations: IGSO3, min_b=1.5,max_b=2.5,min_sigma=0.02,max_sigma=1.5
min_sigma = 1.5
max_sigma = 2.5
min_b = 1.5
max_b = 2.5
ts = np.linspace(0.0, 1.0, num_t)
sigma = min_sigma + ts*min_b  + (1/2)*(ts**2)*(max_b - min_b)
```

```python
plt.plot(np.sqrt(sigma**2 - sigma[0]**2), label=r'$\sqrt{\sigma(t)^2 - \sigma(0)^2}$')
plt.legend()
plt.title('Rotation diffusion parameters')
plt.xlabel('t')
```
